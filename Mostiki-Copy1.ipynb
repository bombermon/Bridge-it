```python

# НОЛИКИ - ХОДЯТ ПЕРВЫМИ

# НАЧАЛО БИБЛИОТЕК


# КОНЕЦ БИБЛИОТЕК 

# НАЧАЛО РАБОТА С МАТРИЦЕЙ

matrix = [
   [' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ', ' '],
   [' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ', ' '],
   [' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ', ' '],
   [' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ', ' '],
   [' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ', ' '],
   [' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ', ' '],
   [' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ', ' '],
   [' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ', ' '],
   [' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ']]

for i in range (9):
    for j in range (9):            #i - движение по вертикали, j - движение по горизонтали 
        if ((i % 2) == 0) and ((j % 2) != 0) :
            matrix[i][j] = 'o' 
        
        if ((i % 2) != 0) and ((j % 2) == 0) :
            matrix[i][j] = 'x'
            
# КОНЕЦ РАБОТА С МАТРИЦЕЙ 



# НАЧАЛО ВСПОМОГАТЕЛЬНЫХ ФУНКЦИЙ -----------------------------------

def printMatrix ( matrix ):    #функция вывода 
    print('  ',end="")
    for i in range(9):
        print(i,end=" ")
    print()
    for i in range ( len(matrix) ): 
        print(i, end=" ")
        for j in range ( len(matrix[i]) ): 
            print (matrix[i][j], end=" ")
        print () 

def InputTempCord ():                  #функция ввода координат
    flag = False
    while flag == False :
        PointVert, PointHor = input('Введите координаты в формате - X пробел Y:  ').split()
        PointHor = int(PointHor)
        PointVert = int(PointVert)
        if (PointHor > 0) and (PointVert > 0) and (PointHor < 9) and (PointVert < 9) and (matrix[PointHor][PointVert] == ' '):
            print('Все ок!')
            print(PointHor, PointVert)
            matrix[PointHor][PointVert] = 'ТЕСТ'
            flag = True 
        else:
            print('Неверные значения!')
            flag = False
    return PointHor, PointVert


def InsertBridge (X, Y, Turn):
    if(matrix[X][Y+1] == 'o'):
        if(Turn == 'игрок нолик'):
            matrix[X][Y] = '-'
        else:
            matrix[X][Y] = '|'
    elif(matrix[X][Y+1] == 'x'):
        if(Turn == 'игрок крестик'):
            matrix[X][Y] = '-'
        else:
            matrix[X][Y] = '|'
            
           

# КОНЕЦ ВСПОМОГАТЕЛЬНЫХ ФУНКЦИЙ -----------------------------------------


# НАЧАЛО ОСНОВНАЯ ФУНКЦИЯ ------------------------------------------------


def main():
    Counter = 0
   
    while True:
        
        if(Counter % 2 == 0):
            Turn = 'игрок нолик'
        else:
            Turn = 'игрок крестик'
            
        printMatrix(matrix)
        
        print('////////////////////////////')
        print('Ходит ',Turn)
        print('////////////////////////////')
        
        
        X , Y = InputTempCord() 
        InsertBridge(X,Y,Turn)

        
        
        Counter += 1

    
# КОНЕЦ ОСНОВНАЯ ФУНКЦИЯ ------------------------------------------------



main()
```
